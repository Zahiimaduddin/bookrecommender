```python
def textual_representation(row):
    textual_representation = f"""Title: {row['title']}
Authors: {row['authors']}
Description: {row['description']}
Categories: {row['categories']}
Published Year: {row['published_year']}
Average Rating: {row['average_rating']}
Number of Pages: {row['num_pages']}
""" 

    return textual_representation

    # f string lets us put stuff in the string
    # tripple quotation means the string saves the line breaks
    # So basically for each horizontal entry, this function takes all the data and formats it into one
    # multi-line string
    
```

```python
print(df.iloc[:5].apply(textual_representation, axis=1).values[0])
```

```python
# This creates a new column in the data frame, then populates it with the string
# we formatted above
df['textual_representation'] = df.apply(textual_representation, axis=1)
```
